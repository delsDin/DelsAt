# Rapport d'Analyse Climatique : Station 65344 (2015-2025)

## 1. Introduction et Méthodologie

Ce projet visait à analyser l'évolution climatique de la région de Cotonou sur une décennie. Le pipeline de données a été conçu pour charger, nettoyer (gestion des valeurs aberrantes et manquantes) et transformer les relevés bruts en indicateurs de tendance.


## 2. Analyse des Températures et Anomalies

Les statistiques descriptives révèlent une température moyenne globale de 27,9°C.

    Détection des anomalies : Le script a identifié 23 jours d'anomalies thermiques extrêmes. Ces jours correspondent à des températures maximales (Tmax​) dépassant le seuil de la moyenne plus deux écarts-types.

    Observation : Ces pics de chaleur se concentrent majoritairement sur les mois de mars et avril, confirmant une intensification de la chaleur juste avant la grande saison des pluies.

## 3. Interprétation des Outliers (Valeurs Aberrantes)

Le graphique Boxplot a mis en lumière des points isolés sous les moustaches (outliers froids).

    Analyse : Contrairement aux erreurs de capteurs, ces points sont des anomalies réelles. Ils traduisent des chutes de température brusques souvent liées à l'Harmattan en saison sèche longue ou à des fronts orageux violents.

    Stabilité : La "Grande Saison des Pluies" se distingue par l'absence d'outliers, témoignant d'une plus grande régularité thermique malgré les précipitations.

## 4. Cycle des Précipitations

L'analyse bicolore (hue) des précipitations confirme un régime bimodal :

    Pics principaux : Juin (Grande saison) et Octobre (Petite saison).

    Saisonnalité : Les mois de décembre à février marquent un stress hydrique prononcé avec des moyennes proches de 0 mm.

## 5. Conclusion Générale

L'analyse sur 10 ans démontre une stabilité des cycles saisonniers, mais la présence de 23 anomalies de chaleur et la variabilité des températures en saison sèche suggèrent une sensibilité accrue aux événements météorologiques extrêmes. Ce tableau de bord permet désormais un suivi précis pour l'anticipation des risques climatiques locaux.
